# INSIGHTS FROM THE WE RATE DOG TWEETER PAGE ANALYSIS

Historical data about dog ratings tweets from the 'we rate dogs' twitter page were gathered through their tweets archive and twitter api, it was then cleaned, tidied and saved as a master dataset with the most relevant information about each tweet, the dog featured in them, the dog's rating and tweet engagement between november 2015 and august 2017 in a well detailed format.

An analysis was then run on the wrangled data that produced the insights shown and briefly explained below:

Insight 1 - According to the results obtained by getting the value counts of the dog predictions from our dataset we get to know that the most popular dog breed according to the dataset is the golden_retriever profusely followed by the Labrador_retriever.

Insight 2 - Using the pandas describe method we see clearly that the average dog rating is about **12.28** or approximately **12.3**. 

Insight 3 - According to the ratings column value counts the highest rating gave by a user is **1776**.

Insight 4 - Using the counts method on the ratings we get to know that the most rated value is **12** with a count of **450**.
